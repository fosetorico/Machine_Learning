```python
import warnings
import pandas as pd
import numpy as np

# ignore all warnings
warnings. filterwarnings("ignore")

# read data into dataframe variable df
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Maze/Rain Prediction/weatherAUS.csv")
print('Size of weather data frame is :', df.shape)

# display first few rows of data
df.head()
```

```python
# give basic statistics of each column
df.describe()
```

```python
col_names = df.columns
col_names
```

```python
#sort in ascending order
df.count().sort_values()
```

```python
#drop columns with significant amount of missing data
#RISK_MM is removed since its reveals the amount of next day rainfall (Data Leakage)
#axis = 1, identifies that we are removing Columns; axis = 0, for removing Rows
# df = df.drop(columns=['Sunshine','Evaporation','Cloud3pm','Cloud9am', 'RISK_MM','Date'],axis=1)
df = df.drop(columns=['Sunshine','Evaporation','Cloud3pm','Cloud9am', 'RISK_MM','Date'],axis=1)
df.shape
```

```python
#remove rows with missing values or NA
#we can do this because we have significant amount of data
df = df.dropna()
df.shape
```

```python
from sklearn import preprocessing

#list of numerical columns; dtype = data type
numerical = [var for var in df.columns if df[var].dtype=='float64']

for col in numerical:
    df[col] = preprocessing.scale(df[col])

df.head()
```

```python
# scipy is not necessary here since we have already scaled our data but provides an alternative way to scale data
from scipy import stats

#Z-score (Standardization): This method scales the data to have a mean (Î¼) of 0 and a standard deviation (Ïƒ) of 1.
#useful when the features follow a normal distribution
z = np.abs(stats.zscore(df._get_numeric_data()))
# print(z.head())

#if all the values within a row is < 3, keep it; else remove it
df= df[(z < 3).all(axis = 1 )]
print(df.shape)
```

```python
#for categorical data

categorical = [var for var in df.columns if df[var].dtype=='object']
print("Number of categorical variables: ", len(categorical))
print(categorical)
```

```python
df['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)
```

```python
categorical_columns = ['Location','WindGustDir', 'WindDir3pm', 'WindDir9am']


# print unique values in each column
for col in categorical_columns:
    print(np.unique(df[col]))

#apply one hot encoding
df = pd.get_dummies(df, columns=categorical_columns)
```

```python
df.head()
```

```python
df[["Location_Adelaide", "Location_Albury", "Location_AliceSprings", "WindGustDir_NE", "WindDir3pm_W"]].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

#use all the variable apart from Raintomorrow
#the model is able to figure out which variables dont have an effect on the prediction and it will assign
#a small parameter value to the variable so it does have a significant impact on the ouput

# set X to all features, set y to our target RainTomorrow
X = df.loc[:, df.columns!='RainTomorrow']
y = df.RainTomorrow
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

logReg = LogisticRegression()
logReg.fit(X_train,y_train)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean

cv = KFold(n_splits=10, random_state=1, shuffle=True)

#cross_val_score, test data evaluation
scores = cross_val_score(logReg, X, y, scoring='accuracy', cv=cv) # cv = cross validation
average_score = mean(scores)

print('Overall Accuracy:', average_score)
```
