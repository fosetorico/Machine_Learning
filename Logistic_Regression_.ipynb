```python
# Read the data
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Maze/Logistic Regression/weatherAlbury.csv")
print('Size of weather data frame is :',df.shape)
df.head()
```

```python
# Preprocess the data

df = df.dropna()
print("new shape:" ,df.shape)
df['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)
```

```python
df.head(20)
```

```python
# Implement Logistic Regression Model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


X = df[["MinTemp", "MaxTemp", "Rainfall", "Humidity9am", "Humidity3pm", "Pressure9am",
        "Pressure3pm", "Temp9am", "Temp3pm", "RainToday"]]
y = df.RainTomorrow

# random_state = number of shuffle
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, random_state=42)

# Default max-iter = 100
logreg = LogisticRegression(max_iter = 200)
logreg.fit(X_train,y_train)

y_pred = logreg.predict(X_test)
# y_pred
```

```python
# Evaluate the model

from sklearn.metrics import accuracy_score

score = accuracy_score(y_test,y_pred)
print('Accuracy :',score)
```

The model get at least 86% of predictions correctly.

```python
# Making predictions
Obs = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Maze/Logistic Regression/weatherObs.xlsx")

Obs['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
newData = Obs.values

y_pred = logreg.predict(newData)
print(y_pred)
```
